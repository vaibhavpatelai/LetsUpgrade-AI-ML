```python
#importing the required libraries
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
from sklearn import neighbors
```

```python
ds = pd.read_csv("//Users/VAIBHAV/Vaibhav/Python/LetUpgrade/LetsUpgradeAssignment/Day27/train.csv")
```

```python
ds.columns
```

```python
ds.isnull().sum()
```

```python
ds = ds.drop(["PassengerId","Name","Ticket","Fare","Cabin"],axis=1)
```

```python
ds.head()
```

```python
le = preprocessing.LabelEncoder()
ds["Sex"] = le.fit_transform(ds["Sex"])
ds["Embarked"] = le.fit_transform(ds["Embarked"])
```

```python
ds.head()
```

```python
Y = ds["Pclass"]
X = ds.drop(["Pclass"],axis=1)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=0)
```

```python
row_count = X_test["Age"].count()
```

```python
for i in range(1,row_count):
    knn = neighbors.KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,Y_train)
    print("For neighbor value -",i)
    print("Accuracy Score:\n",knn.score(X_test,Y_test))
    y_pred = knn.predict(X_test)
    print("Confusion Matrix\n",confusion_matrix(Y_test,y_pred),"\n")
    i = i + 1
```
