__1. Correlation between Attrition and Distance from Home__

H0: There is no significant relation between the Distance From Home and attrition

Ha: There is significant relationship between the Distance From Home and attrition

```python
stats,p = pearsonr(ds.Attrition,ds.DistanceFromHome) #using pearsonr function to find correlation
print(stats,p)
```

Interpretation: Here, p>0.05 means, null hypothesis is
accepted. 

There is no dependancy on Attrition and Distance from home
